```python
from PIL import Image
import pytesseract
import inspect
import cv2
import re
```

```python
webcam = cv2.VideoCapture(0)
while True:
    (_,im) = webcam.read()
    gray = cv2.cvtColor(im,cv2.COLOR_BGR2GRAY)
    #(thresh, img) = cv2.threshold(gray, 128, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)
    img = cv2.threshold(img, 114, 255, cv2.THRESH_BINARY)[1]
    image = Image.fromarray(img)
    text = pytesseract.image_to_string(image)
    text=text.lower()
    name=""
    enroll=""
    try:
        name = re.findall("\nname.+",text)[0].replace('\nname','')
        name = re.findall('[a-z\s]+',name.strip())[0].strip()
        enroll = re.findall("[\d]{11}",text)[0]
    except IndexError:
        print("error",text)
    cv2.putText(im,'{},-{}'.format(name,enroll),(10,10),cv2.FONT_HERSHEY_PLAIN,1,(0,255,0))
    cv2.imshow("I-Card",im)
    key = cv2.waitKey(10)
    if key == 27:
        break
webcam.release()
cv2.destroyAllWindows()
```
